```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(40)
x = np.random.rand(50,1) * 10
y = 3*x + np.random.randn(50,1) * 2

df = pd.DataFrame({
    'input' : x.flatten(),
    'output': y. flatten()
})

df.head()
```

```python
plt.scatter(df['input'], df['output'])
plt.show()
```

```python
m = 0
b = 0
learning_rate = 0.01
```

```python
for i in range(1000):
    y_pred = m * x + b 
    loss = np.mean((y-y_pred)**2)

    dm = (-2/len(x))*np.sum(x*(y-y_pred))
    db = (-2/len(x)) * np.sum((y-y_pred))

    m = m - learning_rate * dm
    b = b - learning_rate * db

    if i%100 == 0:
        print("iteration-",i, "m = ",m, "b = ",b, "loss = ", loss)
    

test = 5
y_prediction = m * test + b

print("For test data ", test,"the prediction is: ", y_prediction)


```
